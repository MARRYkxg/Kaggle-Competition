<div class="alert alert-block alert-success">
<font color='darkblue'>Skew:  </font>

To avoid the long-tail distribution or overdispersion happened, we decided to use Skew to make our data more normal distribution.</font>

```python
from scipy.stats import skew
```

```python
# Numerical Features
num_attr = full_data.dtypes[full_data.dtypes != "object"].index
skewed_attr = full_data[num_attr].apply(lambda x: skew(x.dropna()))
skewed_attr = skewed_attr[skewed_attr > 0.75]
skewed_attr = skewed_attr.index

full_data[skewed_attr] = np.log1p(full_data[skewed_attr])
```
